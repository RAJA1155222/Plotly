Q1

Load the Titanic dataset using seaborn and plot a scatter plot (age vs fare) using Plotly Express

```python
import seaborn as sns
import plotly.express as px

# load titanic dataset
titanic = sns.load_dataset("titanic")

# scatter plot
fig = px.scatter(
    titanic,
    x="age",
    y="fare",
    title="Scatter Plot of Age vs Fare (Titanic)"
)

fig.show()
```

Q2

Using the tips dataset in Plotly, plot a box plot

```python
import plotly.express as px

tips = px.data.tips()

fig = px.box(
    tips,
    x="day",
    y="total_bill",
    title="Box Plot using Tips Dataset"
)

fig.show()
```

Q3

Using the tips dataset, plot a histogram with:

x = "sex"

y = "total_bill"

pattern_shape = "smoker"

color = "day"

```python
import plotly.express as px

tips = px.data.tips()

fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Histogram of Total Bill by Sex"
)

fig.show()
```

Q4

Using the iris dataset, plot a scatter matrix plot

Only use:

sepal_length

sepal_width

petal_length

petal_width

Use species for color

```python
import plotly.express as px

iris = px.data.iris()

fig = px.scatter_matrix(
    iris,
    dimensions=[
        "sepal_length",
        "sepal_width",
        "petal_length",
        "petal_width"
    ],
    color="species",
    title="Scatter Matrix Plot for Iris Dataset"
)

fig.show()
```

Q5

What is Distplot? Plot a distplot using Plotly

✔ What is a Distplot?

A distplot (distribution plot) is used to show:

the distribution of a numerical variable

usually combines a histogram and a density curve (KDE)

Distplot using Plotly

```python
import plotly.figure_factory as ff
import numpy as np

# sample data
data = np.random.randn(100)

fig = ff.create_distplot([data], ["Distribution"])
fig.show()
```
